```python
import requests
import pandas as pd

url = "https://raw.githubusercontent.com/sthemonica/alura-voz/refs/heads/main/Dados/Telco-Customer-Churn.json"

response = requests.get(url)
data = response.json()

df = pd.DataFrame(data)

df
```

```python
import pandas as pd
df_plano = pd.json_normalize(data)
df_plano
```

```python
#Alterando a coluna "account.Charges.Total" para Float

df_plano['account.Charges.Total'] = pd.to_numeric(df_plano['account.Charges.Total'], errors='coerce')
print(df_plano['account.Charges.Total'].dtype)
```

```python
# Eliminando as colunas de Churn

df_plano = df_plano[df_plano['Churn'].str.strip() != '']
print("Número de filas después de eliminar las vacías en 'Churn':", len(df_plano))
```

```python
# Vazios ou em branco
df_plano.apply(lambda x: x.astype(str).str.strip() == '').sum()
```

```python
df_plano = df_plano.drop('Total.Day', errors='ignore')
```

```python
# Calculando os gastos diários
df_plano['Total.Day'] = ((df_plano['account.Charges.Total'] / df_plano['customer.tenure']) /30 ).round(2)

# Inserindo a coluna na posição 19
df_plano.insert(19, 'Total.Day', df_plano.pop('Total.Day'))
```

```python
# Agrupar por tenure e Churn
df_grouped = df_plano.groupby(['customer.tenure', 'Churn']).size().unstack(fill_value=0)

# Calcular a % de churn por tenure
df_grouped['churn_rate'] = (df_grouped.get('Yes', 0) / df_grouped.sum(axis=1)) * 100

# Mostrar o resultado
print(df_grouped[['churn_rate']])
```

```python
# =============================================================================
# 1. REMOÇÃO DE COLUNAS IRRELEVANTES
# =============================================================================
# Identificamos colunas que não contribuem para a previsão
cols_to_drop = ['customerID']  # Exemplo de identificador único
df_plano = df_plano.drop(columns=cols_to_drop, errors='ignore')

# =============================================================================
# 2. TRATAMENTO DE VARIÁVEIS CATEGÓRICAS
# =============================================================================
from sklearn.preprocessing import OneHotEncoder

# a. Converter target 'Churn' para numérico
df_plano['Churn'] = df_plano['Churn'].map({'Yes': 1, 'No': 0})

# b. Listar colunas categóricas
categorical_cols = [
    'customer.gender', 'account.PhoneService', 'account.MultipleLines',
    'account.InternetService', 'account.OnlineSecurity', 'account.OnlineBackup',
    'account.DeviceProtection', 'account.TechSupport', 'account.StreamingTV',
    'account.StreamingMovies', 'account.Contract', 'account.PaymentMethod'
]

# c. Aplicar one-hot encoding
encoder = OneHotEncoder(drop='first', sparse_output=False)
encoded_cols = encoder.fit_transform(df_plano[categorical_cols])
encoded_df = pd.DataFrame(
    encoded_cols,
    columns=encoder.get_feature_names_out(categorical_cols)
)

# d. Combinar com dados originais
df_processed = pd.concat([
    df_plano.drop(columns=categorical_cols),
    encoded_df
], axis=1)

# =============================================================================
# 3. ANÁLISE DE DESBALANCEAMENTO DE CLASSES
# =============================================================================
import matplotlib.pyplot as plt

# Calcular proporção de churn
churn_rate = df_processed['Churn'].value_counts(normalize=True)
print(f"\nDistribuição de Churn:\n{churn_rate}")

# Visualizar desbalanceamento
plt.figure(figsize=(8, 4))
churn_rate.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Proporção de Clientes que Deram Churn')
plt.xticks([0, 1], ['Não Churn (0)', 'Churn (1)'], rotation=0)
plt.ylabel('Proporção')
plt.show()

# =============================================================================
# 4. PREPARAÇÃO PARA MODELAGEM
# =============================================================================
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE

# a. Separar features e target
X = df_processed.drop(columns=['Churn', 'Total.Day'])  # Remover possíveis vazamentos
y = df_processed['Churn']

# b. Dividir dados
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# c. Normalizar dados (opcional - necessário para alguns modelos)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# d. Aplicar SMOTE para balanceamento (se necessário)
if churn_rate[1] < 0.3:  # Se churn < 30% dos dados
    smote = SMOTE(random_state=42)
    X_train_bal, y_train_bal = smote.fit_resample(X_train_scaled, y_train)
    print(f"\nDistribuição após SMOTE: {pd.Series(y_train_bal).value_counts(normalize=True)}")
else:
    X_train_bal, y_train_bal = X_train_scaled, y_train.copy()
```
