Random Forest: Modelo de aprendizado baseado em árvores de decisão, altamente interpretável e eficaz em lidar com dados desbalanceados.

Regressão Logística: Modelo linear que estima a probabilidade de um evento (evasão) ocorrer com base em variáveis independentes.
